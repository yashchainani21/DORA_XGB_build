The goal of this notebook is to explore the labelling of processed reactions using thermodynamics only. All reactions are labelled by setting a critical threshold for the reaction maximum/ minimum driving force (MDF) value or simply, the minimum change in Gibbs free reaction energy that can be achieved across a range of metabolite concentrations (subject to cofactor ratios). 

Reactions with MDF values that fall below the predetermined critical threshold are labelled as being thermodynamically infeasible (recall that MDF has the opposite sign to dG) while reactions with MDF values that fall above the predetermined critical threshold are labelled as being thermodynamically feasible.

### <a id="section_3"></a>3. Examining feasible/ infeasible monooxygenase reactions

```python
with open('../../data/processed/all_processed_rxns_with_thermo_labelled_crit_DF_10.pkl','rb') as file:
    labelled_BKM_rxns = pickle.load(file)
```

```python
pos_rxns_by_thermo_count = 0
neg_rxns_by_thermo_count = 0
for rxn in labelled_BKM_rxns:
    if rxn['remark'] == 'pos rxn by thermo':
        pos_rxns_by_thermo_count += 1
    if rxn['remark'] == 'neg rxn by thermo':
        neg_rxns_by_thermo_count += 1
```

```python
pos_rxns_by_thermo_count
```

```python
neg_rxns_by_thermo_count
```

```python
# initialize a counter to keep track of reactions that are feasible in one direction only (i.e. irreversible)
irreversible_rxns_count = 0 

# initialize a counter to keep track of reactions that are feasible in both directions (i.e. reversible)
reversible_rxns_count = 0

# initialize a counter to keep track of the total number of positive reactions parsed
num_rxns_parsed = 0

pos_rxns = [rxn for rxn in labelled_BKM_rxns if rxn['remark'] == 'pos rxn by thermo']


for pos_rxn in pos_rxns:
    
    num_rxns_parsed +=1 
    print(f'\nNumber of reactions parsed: {num_rxns_parsed}')
    
    substrates = set(pos_rxn['Substrates'])
    products = set(pos_rxn['Products'])
    lhs_cofactors = set(pos_rxn['LHS_cofactors'])
    rhs_cofactors= set(pos_rxn['RHS_cofactors'])
    
    # now, we find the reverse of this particular positive reaction in the list of all labelled reactions
    for rxn in labelled_BKM_rxns:
        if set(rxn['Products']) == substrates and set(rxn['Substrates']) == products and set(rxn['LHS_cofactors']) == rhs_cofactors and set(rxn['RHS_cofactors']) == lhs_cofactors:
            
            print('reverse found')
            # once the reverse reaction has been found, if this reverse is also feasible, 
            # then this reaction is feasible in both directions, i.e. it is a reversible reaction
            
            if rxn['remark'] == 'pos rxn by thermo':
                reversible_rxns_count += 1
        
            # but if this reverse is not feasible,
            # then this reaction was feasible in one direction only, i.e. it is an irreversible reaction
            if rxn['remark'] == 'neg rxn by thermo':
                irreversible_rxns_count += 1
```
